```python
# today will sove a problem , like exam mark grading

marks = float(input("Enter your marks: "))
if marks >=90 and marks <=100:
  print('your grade is A+')

elif marks >=80 and marks <90:
  print('your grade is : A-')
elif marks >=70 and marks <80:
  print('your grade is : B')

elif marks >=60 and marks <70:
  print('your grade is : passed')
elif marks <60:
  print('your grade is : failed')
else:
  print('invalid marks')
```

```python
# so lets solve another simple problem with if else condition

num1 = int (input("enter num1:"))
num2 = int (input("enter num2:"))
num3 = int 3(input("enter num3:"))
print('your inputed number is :', num1,",", num2,",", num3)

if num1 > num2 and num1 > num3:
  print('The greater is : ', num1)

elif num2 > num1 and num2 > num3:
  print('The greater num is :', num2)

else:
  print('The greater num is :', num3)
```
